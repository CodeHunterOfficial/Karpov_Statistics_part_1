```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
url = 'http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv'
df = pd.read_csv(url)
df
```

```python
df[['hs_grad', 'poverty']].describe()
```

```python
sns.scatterplot(data=df, x=df['hs_grad'], y=df['poverty'], color='orange');
print('Диаграмма рассеивания')
```

```python
r_xy = np.corrcoef(x=df['hs_grad'], y=df['poverty'])[0, 1]
print(f"Коэффициент корреляции равен {r_xy}")
```

```python
R_sq = r_xy ** 2
print(f'Коэффициент детерминации равен {R_sq}')
```

```python
y = df['poverty'].to_numpy().reshape(-1, 1)
X = df['hs_grad'].to_numpy().reshape(-1, 1)

LR = LinearRegression()
model = LR.fit(X, y)
```

```python
b0 = model.intercept_[0]
b0 = round(b0, 4)
b0
```

```python
b1 = model.coef_[0][0]
b1 = round(b1, 4)
b1
```

```python
Y = [b0 + b1 * i for i in X]
plt.xlabel('Уровень образования')
plt.ylabel('Уровень бедности')
plt.title('Линейная регрессия')
plt.scatter(X, y);
plt.plot(X, Y);
```

```python
LR_equation = f'y = {b0} + {b1}X' if b1 > 0 else f'y = {b0} + ({b1})X'
print(f'Уравнение линейной регрессии: {LR_equation}')
```
